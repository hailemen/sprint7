Exercici 3. Comptador de paraules d'un text.

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula.

```python
# Versión antigua(NO SE PRESENTA): Defino esta función que  me permite remover todos los signos de puntuación que encuentre en el texto para poder contarlos correctamente.
def remover_puntuacion(texto):
    signos_puntuacion = [".", ",", "?", "¿", "!", "¡", "(", ")", ";", ":", "-", "'", '"']
    for signo in signos_puntuacion:
        texto = texto.replace(signo, "") # Reemplazamos el signo por 'nada' para que lo borre
    return texto
     
def contador_veces(texto):
    texto = remover_puntuacion(texto)
    texto = texto.lower() # Para convertir todo el texto en minúsculas y no cuente por ejemplo "Hola" y "hola" como palabras separadas
    lista_palabras = texto.split()  # Para separar las palabras del texto
    frecuencia = {}     # Se crea un diccionario vacío                 
    for palabra in lista_palabras:
        frecuencia[palabra] = lista_palabras.count(palabra)  # Al diccionario se le va agregando el conteo de las palabras de la lista
    return frecuencia

texto = input('Escribe un texto: ')
resultado = contador_veces(texto)

for palabra,numero in resultado.items():
    print(f'{palabra}:{numero}')
```

```python
# EJERICIO 3. VERSIÓN NUEVA (La optimización se ha hecho en el for indicado en correción. Se elimina el contador utilizado por un método que recorre la lista una sola vez,
# donde iterará solo en el diccionario)

# Defino esta función que  me permite remover todos los signos de puntuación que encuentre en el texto para poder contarlos correctamente.
def remover_puntuacion(texto):
    signos_puntuacion = [".", ",", "?", "¿", "!", "¡", "(", ")", ";", ":", "-", "'", '"']
    for signo in signos_puntuacion:
        texto = texto.replace(signo, "") # Reemplazamos el signo por 'nada' para que lo borre
    return texto
     
def contador_veces(texto):
    texto = remover_puntuacion(texto)
    texto = texto.lower() # Para convertir todo el texto en minúsculas y no cuente por ejemplo "Hola" y "hola" como palabras separadas
    lista_palabras = texto.split()  # Para separar las palabras del texto
    frecuencia = {}     # Se crea un diccionario vacío                 
    for palabra in lista_palabras:
        if palabra in frecuencia:  
            frecuencia[palabra] +=1  # Cambio realizado: en la función si encuentra la palabra en el diccionario suma al diccionario frecuencia
        else:
            frecuencia[palabra]= 1  # Cambio realizado: Si no lo encuentra, solo será uno. El recorrido de la función se reduce con este método
    return frecuencia

texto = input('Escribe un texto: ')
resultado = contador_veces(texto)

for palabra,numero in resultado.items():
    print(f'{palabra}:{numero}')
```

Diccionari invers.

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, 
és a dir, intercanviats les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas,
la funció hauria d'imprimir un missatge d'advertiment.

```python
# EJERCICIO 4. VERSIÓN NUEVA (SE PRESENTA). Modificado: El manejo del mensaje de error es más claro, y solo devuelve el diccionario nuevo si no hay duplicados, en caso contrario manda el error indicado

# Diccionario con advertencia de duplicados (Si en el diccionario antiguo hay una clave duplicada no aparecerá)

diccionario_antiguo = {"a": 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5, 'f': 6} 

def reordenar1(diccionario_antiguo):
    
    diccionario_nuevo = {} # Se crea el diccionario vacío
      
    for clave, valor in sorted(diccionario_antiguo.items() ):
        if valor in diccionario_nuevo: # Si encuentra un valor duplicado interrumpe el bucle y lanza el mensaje de error
            print('Error: multiple keys for one value') 
            return
        else:                           # Si no lo encuentra lo agrega al diccionario vacío, invirtiendo el valor que será ahora la clave
            diccionario_nuevo[valor] = clave
    print(f'El diccionario nuevo es: {(diccionario_nuevo)}') 
    
reordenar1(diccionario_antiguo)
```

Continuant amb l'exercici 4 del nivell 1: al client es va oblidar de comentar un detall i resulta que els valors en el diccionari original poden duplicar-se i més, per la qual cosa les claus intercanviades poden tenir duplicats. En aquest cas, en l'exercici anterior imprimies un missatge d'advertiment, ara, els valors del diccionari resultant hauran d'emmagatzemar-se com una llista. Tingues en compte que si és un valor únic no ha de ser una llista.

```python
# NO HAY CORRECCIONES SOLICITADAS. SE MANTIENE

diccionario_antiguo = {"a": 1, 'b': 2, 'c': 3, 'd': 4, 'e': 3 }
diccionario_nuevo = {} 

def reordenar2(diccionario_antiguo):
    
    for clave, valor in diccionario_antiguo.items():   # Se usan las función items para iterar sobre cada par (clave-valor)
        if valor not in diccionario_nuevo:         # la función if se utiliza para saber si el valor nuevo ya existe como clave  
            diccionario_nuevo[valor] = clave       # Si no existe crea una nueva lista vacía usándolo como valor asociado y se agrega la clave antigua a esa lista  
        else:
            if isinstance(diccionario_nuevo[valor], list):
                diccionario_nuevo[valor].append(clave)
            else:
                diccionario_nuevo[valor] = [diccionario_nuevo[valor], clave]   # Si existe, se obtiene la lista existente al valor nuevo se agrega la clave antigua a la lista existente
    return diccionario_nuevo

reordenar2(diccionario_antiguo)

print(f'El diccionario antiguo era: {diccionario_antiguo}')
print(f'El diccionario nuevo es: {diccionario_nuevo}')
```

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir

Exemple de la llista que rep el client: [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

En los ejercicios de prueba utilicé esta lista que tiene un dato de cada tipo: entero, float, string, complejo, diccionario, tupla y lista para saber que funcionaba
[2+2, 1, 5.2, "Hola", 3j, {"a": 1}, (2, 3), [4, 5, (8,9)]] Que después uní con la lista de ejemplo para probar el conjunto.

```python
# EJERCICIO 2. CON LAS MOFICIACIONES SOLICITADAS, SE ELIMINA LOS ELEMENTOS DE LA LISTA QUE NO FUERON SOLICITADOS EN EL ENUNCIADO.

def conversor(dato, lista_previa, lista_convertibles, lista_no_convertibles):
    # Verificar si el dato es una estructura de datos compuesta
    if isinstance(dato, dict):
        # Iterar sobre pares clave-valor
        for k, v in dato.items():
            lista_previa.append(v)  # Añadir la clave a lista_previa
            conversor(k, lista_previa, lista_convertibles, lista_no_convertibles)  # Recursivamente procesar la clave
            conversor(v, lista_previa, lista_convertibles, lista_no_convertibles)  # Recursivamente procesar el valor
    elif isinstance(dato, (list, tuple)):
        for item in dato:
            conversor(item, lista_previa, lista_convertibles, lista_no_convertibles)
    else:
        lista_previa.append(dato)  # Añadir el dato a lista_previa
        try:
            float_dato = float(dato)
            lista_convertibles.append(float_dato)
        except (ValueError, TypeError):
            lista_no_convertibles.append(dato)

# Aquí se encuentra la función donde primero se crean las listas vacias y se almacenan los datos cuando se itere sobre la lista base, la previa con los datos originales,
# una lista con los datos que son convertibles y otra con los datos que no se pueden convertir. Finalmente se imprimen.

def tratamiento(lista_base):
    lista_previa = []
    lista_convertibles = []
    lista_no_convertibles = []
    
    for elemento in lista_base:
        conversor(elemento, lista_previa, lista_convertibles, lista_no_convertibles)

    print('La lista previa es:', lista_previa)
    print('Números convertibles a flotantes:', lista_convertibles)
    print('Números no convertibles a flotantes:', lista_no_convertibles)

lista_base = ['1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']] 
tratamiento(lista_base)
```

```python
import re
import json

def split_text_into_words(text):                # Esta función divide el texto dado en palabras y lo convierte a minúsculas.
    return re.findall(r"[\w']+", text.lower())

nombre_archivo = "tu_me_quieres_blanca.txt"          # Lee el contenido del archivo "tu_me_quieres_blanca.txt"
with open(nombre_archivo, 'r', encoding='utf8') as archivo:
    contenido = archivo.read()

palabras = split_text_into_words(contenido)         

def contador_palabras(palabras):   # Esta función crea un diccionario para contar las ocurrencias de cada palabra
    alfabet = {}                      
    for palabra in palabras:
        kw = palabra[0]
        if kw not in alfabet.keys():
            alfabet.update({kw:{}})
        if palabra not in alfabet.values():
            recuento = palabras.count(palabra)
            alfabet[kw] [palabra] = recuento
    return alfabet

def ordenador_diccionario(diccionario):   # Esta función ordena el diccionario por clave y luego ordena cada valor interno.
    diccionario_ordenado = {}
    for kw in sorted(diccionario):
        diccionario_ordenado[kw] = {}
        for interkw in sorted(diccionario[kw]):
            diccionario_ordenado[kw][interkw] = diccionario[kw][interkw]
    print(json.dumps(diccionario_ordenado, indent=2, ensure_ascii=False)) # Se imprime la representación formateada y legible del diccionario
                                                                          # con dos espacios de sangría y se asegura que no escape caracteres no ascii

palabras_ordenadas = contador_palabras(palabras)
ordenador_diccionario(palabras_ordenadas)
```
